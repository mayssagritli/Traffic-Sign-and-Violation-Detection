```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '/content/drive/MyDrive/Training_results/Road-Sign-Detection2/results.csv'
results_df = pd.read_csv(file_path)

# Check and strip column names
results_df.columns = results_df.columns.str.strip()

# Extract the relevant data
epochs = results_df['epoch']
train_box_loss = results_df['train/box_loss']
val_box_loss = results_df['val/box_loss']
train_cls_loss = results_df['train/cls_loss']
val_cls_loss = results_df['val/cls_loss']
train_dfl_loss = results_df['train/dfl_loss']
val_dfl_loss = results_df['val/dfl_loss']
map50 = results_df['metrics/mAP50(B)']
map50_95 = results_df['metrics/mAP50-95(B)']

# Plot the losses and accuracy on the same plot with dual y-axes
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plotting Losses
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss', color='tab:blue')
ax1.plot(epochs, train_box_loss, label='Train Box Loss', color='tab:blue', linestyle='--')
ax1.plot(epochs, val_box_loss, label='Val Box Loss', color='tab:blue')
ax1.plot(epochs, train_cls_loss, label='Train Cls Loss', color='tab:orange', linestyle='--')
ax1.plot(epochs, val_cls_loss, label='Val Cls Loss', color='tab:orange')
ax1.plot(epochs, train_dfl_loss, label='Train DFL Loss', color='tab:green', linestyle='--')
ax1.plot(epochs, val_dfl_loss, label='Val DFL Loss', color='tab:green')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.legend(loc='upper left')

# Create a second y-axis for accuracy
ax2 = ax1.twinx()
ax2.set_ylabel('mAP', color='tab:red')
ax2.plot(epochs, map50, label='mAP50(B)', color='tab:red')
ax2.plot(epochs, map50_95, label='mAP50-95(B)', color='tab:purple')
ax2.tick_params(axis='y', labelcolor='tab:red')
ax2.legend(loc='upper right')

fig.tight_layout()
plt.title('Loss and mAP Curves')
plt.show()
```
